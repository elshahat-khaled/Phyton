2. gli indirizzi in cui si trovano gli impianti dell’ATM


```python
result=db.impianti.find({"Sit_NomeImpianto":{"$regex":"ATM"}},{"Sit_NomeImpianto":1,"_id":0,"INDIRIZZO":1})
df = pd.DataFrame(list(result))
df
```

3.l’elenco degli impianti della Vodafone e della TIM

```python
result=db.impianti.find({"$or":[{"gestore":{"$regex":"VODAFONE"}},{"gestore":{"$regex":"TIM"}}]})
df = pd.DataFrame(list(result))
df
```

4.gli indirizzi in cui c’è più di un impianto


```python
result = db.impianti.aggregate([{ "$group" : {"_id": "$INDIRIZZO", "total" : {"$sum" : 1}}},
                                { "$match": { "total": { "$gt": 1 }}},
                                { "$project": {"_id":1,"total":0}}])
df2 = pd.DataFrame(list(result))
df2
```

5.i gestori che sono società per azioni, Ordinare in ordine alfabetivo.

```python
result=db.impianti.find({'gestore':{ "$regex": "S.p.A." }},{"gestore":1,"_id":0}).sort("gestore",1)
df = pd.DataFrame(list(result))
df
```

6.il numero di impianti per ogni gestore

```python
result = db.impianti.aggregate([{ "$group" : {"_id": "$gestore", "total" : {"$sum" : 1}}}])
df2 = pd.DataFrame(list(result))
df2
```

7.i gestori che hanno piu di 100, ordinare in ordine decrescente a partire dal gestore ce ha piu impianti

```python
result = db.impianti.aggregate([{ "$group" : {"_id": "$gestore", "total" : {"$sum" : 1}}},
                                { "$match": { "total": { "$gt": 100 }}},
                                { "$sort": {"total": -1 }}])
df2 = pd.DataFrame(list(result))
df2
```

11.gli impianti che non sono acessi e chi si trovano a Cimiano e nel Quartiere Adriano

```python
result = db.impianti.find({ "$or":[{"Sit_StatoImpianto" : {'$not': { "$regex": "^Acceso.*" }}},{"NIL":{"$regex":"CIMIANO"}}]})
df2 = pd.DataFrame(list(result))
df2
```

12. il quartiere con più impianti

```python
result = db.impianti.aggregate([{ "$group" : {"_id": "$MUNICIPIO", "total" : {"$sum" : 1}}},
                                {"$sort":{"_id":-1}}])
df2 = pd.DataFrame(list(result))
df2
```

### 1. visualizzi su una mappa i 10 corsi d'acqua più lunghi della provincia di Milano

```python
corsi_milano = Corsi_acqua[Corsi_acqua.within(milano.geometry.squeeze())]
corsi_piu_lun= corsi_milano.nlargest(10, "SHAPE_LEN")
corsi_piu_lun
```

```python
ax = corsi_piu_lun.to_crs(epsg=3857).plot(figsize= (15, 15), edgecolor="k", linewidth=2)
milano.to_crs(epsg=3857).plot(ax=ax, facecolor= "Green",alpha = 0.5, linewidth= 3, edgecolor= "k")
contextily.add_basemap(ax=ax)
```

### 2. visualizzi su una mappa i Navigli e i fiumi da cui prendono l'acqua. Utilizzare colori diversi per rappresentare i navigli rispetto ai fiumi ed evidenziare bene i percorsi sulla mappa

```python
navigli = Corsi_acqua[Corsi_acqua["NOME"].str.contains("Navigli")]
fiumi = Corsi_acqua[Corsi_acqua.intersects(navigli.geometry.squeeze().unary_union)]
fiumi = fiumi[~fiumi["NOME"].str.contains("Navigli")]
```

```python
ax = milano.to_crs(epsg=3857).plot(figsize=(20,20), facecolor= "Green",alpha = 0.5, linewidth= 3, edgecolor= "k")
navigli.to_crs(epsg=3857).plot(ax=ax, linewidth= 3, edgecolor= "b")
fiumi.to_crs(epsg=3857).plot(ax=ax, linewidth= 3, edgecolor= "k")
contextily.add_basemap(ax=ax)
```

3. visualizzi su una mappa i corsi d'acqua che passano in un comune inserito dall'utente

### 4. visualizzi su una mappa un fiume inserito dall'utente con tutti i comuni  che attraversa. Se attraversa Milano, visualizzare solo i quartieri che attraversa

```python
nome_Fiume = input("inserire il nome del Fiume: ")
Fiume = Corsi_acqua[Corsi_acqua.NOME == nome_Fiume]
Fiume
```

```python
ax = Fiume.to_crs(epsg = 3857).plot(figsize=(15,15), facecolor = "none",alpha= 0.5, edgecolor = "k", linewidth = 2)
Fiume_att = Corsi_acqua[Corsi_acqua.intersects(Fiume.geometry.squeeze())]
Fiume_att.to_crs(epsg = 3857).plot(ax=ax, facecolor = "y", edgecolor = "k", linewidth = 4)
contextily.add_basemap(ax)
```

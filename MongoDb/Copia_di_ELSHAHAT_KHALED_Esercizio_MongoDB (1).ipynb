Estrarre dalla collection le informazioni richieste:

```
# gli appartamenti che costano più di 100 $ a notte
appa = db.listingsAndReviews.find({"price": {'$gt' : 100 }})
df_co_al = pd.DataFrame(list(appa))
df_co_al
```

```
# gli appartamenti che costano tra i 100 $ e i 200 $ a notte
appa = db.listingsAndReviews.find({"price": {'$gt': 100, '$lt': 200 } });
costo_100_200 = pd.DataFrame(list(appa))
costo_100_200
```

```
# gli appartamenti che hanno almeno 5 posti letto e che costano tra i 50 $ e 100 $ a notte
appa = db.listingsAndReviews.find({"$and" : [{"price" : {'$gt' : 50, '$lt' : 100}}, {"beds" : {'$gt' : 5 }}]})
posti_letto_5 = pd.DataFrame(list(appa))
posti_letto_5
```

```
# gli appartamenti in Portogallo
appa = db.listingsAndReviews.find({'address.country_code' : 'PT'})
app_porto = pd.DataFrame(list(appa))
app_porto
```

```
# gli appartamenti che non si trovano in Portogallo e in Spagna
appa = db.listingsAndReviews.find({"address.country_code": {'$not' : {"$in": ["PT", "ES"]}}})
non_sp_port = pd.DataFrame(list(appa))
non_sp_port
```

```
# gli appartamenti negli Stati Uniti che hanno il wifi
# ordinare in ordine decrescente sul prezzo per notte e prendere solo i primi 10 appartamenti
appa = db.listingsAndReviews.find({"address.country_code": {"$in": ["US"]}})
appa_usa = pd.DataFrame(list(appa))
appa_usa
```

```
# il numero di appartamenti in Portogallo
# il numero di appartamenti in Portogallo
appa = db.listingsAndReviews.aggregate([ 
                                   { "$match": { "address.country_code": "PT" } }, 
                                   { "$group" : {"_id": "$address.country", "total" : {"$sum" : 1}}} , 
                                    ])
n_app_port = pd.DataFrame(list(appa))
n_app_port
```

```
# il numero di appartamenti nelle varie nazioni
appa = db.listingsAndReviews.aggregate([{ "$group" : {"_id": "$address.country", "total" : {"$sum" : 1}}}])
n_appa_nazioni = pd.DataFrame(list(appa))
n_appa_nazioni
```

```
# le nazioni con più di 600 appartamenti
appa = db.listingsAndReviews.aggregate([{ "$group" : {"_id": "$address.country", "total" : {"$sum" : 1}}},
                                          { "$match": { "total": { "$gt": 600 }}}])
nazioni_piu_600 = pd.DataFrame(list(appa))
nazioni_piu_600
```

```
# gli appartamenti che hanno la parola 'beach' nella descrizione (google it...)
appa = db.listingsAndReviews.aggregate([
                                          { "$match": { "description": {"$regex" : "beach"}}}])
appa_beach = pd.DataFrame(list(appa))
appa_beach
```

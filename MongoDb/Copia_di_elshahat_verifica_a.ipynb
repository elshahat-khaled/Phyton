10.il numero di consolati in ogni continente

```python
result = db.consolati_italiani.aggregate([
                {"$group" : {"_id": "$continente", "total" : {"$sum" : 1}}},
                { "$sort": {"_id": 1 }}
                ])
df = pd.DataFrame(list(result))
df
```

```python
result = db.consolati_italiani.aggregate([{ "$unwind": "$continente" },
                              { "$group" : {"_id": "$continente", "total" : {"$sum" : 1}}},
                              { "$project": {"_id": 0,"continente": "$_id","total": "$total"}}])
df = pd.DataFrame(list(result))
df
```

esercizio 11 i continenti che hanno piÃ¹ di 50 consolati. Ordinare in ordine decrescente a partire dal continente che ha piÃ¹ consolati


```python
result = db.consolati_italiani.aggregate([
                {"$group" : {"_id": "$continente", "total" : {"$sum" : 1}}},
                {"$match":{"total":{"$gt":50}}},
                { "$sort": {"total": -1 }}
                ])
df = pd.DataFrame(list(result))
df
```

```python
result = db.consolati_italiani.aggregate([{ "$unwind": "$continente" },
                              { "$group" : {"_id": "$continente", "total" : {"$sum" : 1}}},
                              { "$match": { "total": { "$gt": 50 }}},
                              { "$sort": {"total": -1 }},
                              { "$project": {"_id": 0,"continente": "$_id","total": "$total"}}])
df = pd.DataFrame(list(result))
df
```

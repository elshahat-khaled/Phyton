7.     
Il numero di film per ogni lingua. Ordinare in ordine decrescente a partire
della lingua più presente

```python
result = db.movies.aggregate([  
                                   { "$unwind": "$languages" }, 
                                   {"$group" : {"_id": "$languages", "total" : {"$sum" : 1}}} , 
                                   { "$sort": { "total": -1 } } 
                                   ])
df = pd.DataFrame(list(result))
df
```

8.     
Le prime 3 lingue in cui sono stati girati i film

```python
result = db.movies.aggregate([  
                                   { "$unwind": "$languages" }, 
                                   {"$group" : {"_id": "$languages", "total" : {"$sum" : 1}}} , 
                                   { "$sort": { "total": -1 } },
                                   {"$limit":3}
                                   ])
df = pd.DataFrame(list(result))
df
```

9.     
Le lingue in cui sono stati girati meno di 50 film

```python
result = db.movies.aggregate([  
                                   { "$unwind": "$languages" }, 
                                   {"$group" : {"_id": "$languages", "total" : {"$sum" : 1}}} , 
                                   { "$sort": { "total": -1 } },
                                   { "$match": { "total": { "$lt": 50 }}}
                                   ])
df = pd.DataFrame(list(result))
df
```

**1. I titoli dei film**

```python
result=db.movies.find({},{"title":1,"_id":0})
df = pd.DataFrame(list(result))
df
```

**2. I film che nel titolo hanno la parola &quot;the&quot;**

```python
result=db.movies.find({"title":{"$regex":"the"}})
df = pd.DataFrame(list(result))
df[['title']]
```

```python
result = db.movies.aggregate([{ "$match": { "title": {"$regex" : "the"}}}])
df = pd.DataFrame(list(result))
df[['title']]
```

**3. I film americani degli anni 20**


```python
result=db.movies.find({'$and':[{"year":{"$gt":1919,"$lt":1930}},{'countries':"USA"}]})
df = pd.DataFrame(list(result))
df
```

**4. I film americani degli anni 30 e degli anni 4**0

```python
result=db.movies.find({'$and':[{"year":{"$gt":1929,"$lt":1950}},{'countries':"USA"}]})
df = pd.DataFrame(list(result))
df
```

```python
result=db.movies.find({'$and':[{'countries':"USA"}, {'$or': [{"year":{"$gt":1929, "$lt":1940}}, {"year":{"$gt":1949, "$lt":1960}}]}]})
df = pd.DataFrame(list(result))
df
```

**5. Tutti i film tranne quelli americani**


```python
result=db.movies.find({"countries":{"$not":{"$regex" :"USA"}}})
df = pd.DataFrame(list(result))
df
```

```python
result=db.movies.find({"countries" : {'$not': {"$eq":"USA"}}},{"title": 1, "year" : 1, "countries": 1,"_id":0})
df=pd.DataFrame(list(result))
df
```

**6. I film che hanno un rating superiore a 4**


```python
result=db.movies.find({"imdb.rating":{"$gt":4}},{'title':1, 'imdb':1,'_id':0})
df = pd.DataFrame(list(result))
df
```

**7. Il numero di film per ogni lingua. Ordinare in ordine decrescente a
partire della lingua più presente**


```python
result = db.movies.aggregate([{ "$unwind": "$languages" },
                              { "$group" : {"_id": "$languages", "total" : {"$sum" : 1}}},
                              { "$sort": {"total": -1 }},
                              { "$project": {"_id": 0,"languages": "$_id","total": "$total"}}])
df2 = pd.DataFrame(list(result))
df2
```

**8. Le prime 3 lingue in cui sono stati girati i film**


```python
result = db.movies.aggregate([{ "$unwind": "$languages" },
                              { "$group" : {"_id": "$languages", "total" : {"$sum" : 1}}},
                              { "$sort": {"total": -1 }},
                              { "$project": {"_id": 0,"languages": "$_id","total": "$total"}},
                              { "$limit" : 3 }])
df2 = pd.DataFrame(list(result))
df2
```

**9. Le lingue in cui sono stati girati meno di 50 film**


```python
result = db.movies.aggregate([{ "$unwind": "$languages" },
                              { "$group" : {"_id": "$languages", "total" : {"$sum" : 1}}},
                              { "$match": { "total": { "$lte": 50 }}},
                              { "$sort": {"total": -1 }},
                              { "$project": {"_id": 0,"languages": "$_id","total": "$total"}}])
df2 = pd.DataFrame(list(result))
df2
```

**10. La durata media dei film italiani**


```python
result = db.movies.aggregate([{"$unwind": "$languages"},
                            {"$match" : {"languages" : "Italian"}},
                            {"$group": {"_id": "$languages", "media": {"$avg":"$runtime"}}}])
df=pd.DataFrame(list(result))
df
```

**11. La durata media dei film delle varie nazioni**


```python
result = db.movies.aggregate([{"$unwind": "$languages"},
                            {"$group": {"_id": "$languages", "media": {"$avg":"$runtime"}}}])
df=pd.DataFrame(list(result))
df
```

**12. Il film che è dura di più**

```python
result = db.movies.aggregate([{"$group": {"_id" : "$runtime"}},
                              {"$sort":{"_id":-1}},
                              {"$limit":1}])
df = pd.DataFrame(list(result))
max_length = df.iloc[0,0]

result=db.movies.find({"runtime":max_length.astype(float)})
df = pd.DataFrame(list(result))
df[['title','runtime']]
```

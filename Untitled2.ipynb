```python
df = pd.merge(dfOrders,dfOrdersItems,how="outer",on=["order_id"])
df["prezzo_totale"]= df["list_price"]*df["quantity"]*(1-df["discount"])
df
```

```python
df1=df.groupby(["store_id","order_date"]).sum()["prezzo_totale"].reset_index()
```

```python
import calendar
df1['year'] = pd.DatetimeIndex(df1['order_date']).year
df1['nMonth'] = pd.DatetimeIndex(df1['order_date']).month
df1['Month'] = df1['nMonth'].apply(lambda x: calendar.month_abbr[x])
```

```python
df1
```

```python
df2 = df1.groupby(['nMonth','year','Month']).sum()#.reset_index()
df2 #[['nMonth','year',	'Month','prezzo_totale']]
```

```python
df2016 = df2.iloc[df2.index.get_level_values('year') == 2016].reset_index()
df2016
```

```python
df2017 =df2.iloc[df2.index.get_level_values('year') == 2017].reset_index()
df2017
```

```python
df2018t =df2.iloc[df2.index.get_level_values('year') == 2018]
df2018 = df2018t.append(pd.Series(name=((5, 2018, 'May'))))
df2018 = df2018.reset_index().sort_values(by="nMonth")
df2018
```

```python
labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

x = np.arange(len(labels))  
width = 0.25  

fig, ax = plt.subplots(figsize = (11,8))
N1 = ax.bar(x-0.25, df2016.prezzo_totale, width,label='2016')
N2 = ax.bar(x, df2017.prezzo_totale, width,label='2017')
N3 = ax.bar(x+0.25, df2018.prezzo_totale, width,label='2018')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_xlabel('Scores')
ax.set_title('Scores by group and gender')

#ax.set_xticklabels(x + width/2,labels)
plt.xticks(x,labels)
ax.legend()

fig.tight_layout()
plt.show()
```
